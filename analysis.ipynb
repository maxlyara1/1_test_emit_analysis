# Анализ распределения оценок за первую контрольную работу по макроэкономике по группам

## Импортируем нужные библиотеки и файл

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Микроэкономика 2022-2023 РАНХиГС ЭМИТ 2-семестр.csv')
```

```python
df
```

## Оставляем только нужные для анализа столбцы

```python
dfc = df.drop(columns=['Unnamed: 1', 'Комментрарий 1', 'Комментрарий 2', 'Комментрарий 3', 'Комментрарий 4','Комментарии проверяющих','Оценка за КР 1 из 10 баллов'])
```

## Из интересного можно отметить корелляцию между баллами за 4 задание и за всю работу

```python
plt.figure(figsize=(12,8), dpi= 100)
sns.pairplot(dfc, kind="scatter", hue='Группы')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,8), dpi= 100)
sns.heatmap(dfc.corr(), xticklabels=dfc.corr().columns, yticklabels=dfc.corr().columns, center=0, annot=True, cmap = 'Blues')

# Decorations
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

## Замечаем у группы П-22-101 ненормальное распределение оценок за конрольную работу, что не является нормой. Возможно это из-за небольшого количества данных об этой группе.

```python
# Draw Plot
plt.figure(figsize=(16,10), dpi= 500)
sns.violinplot(x='Группы', y='Итого КР 1', data=df, scale='width', inner='quartile')

# Decoration
plt.title('Распределение оценок за КР по макре', fontsize=14)
plt.show()
```
